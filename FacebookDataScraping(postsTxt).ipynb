```python
from facebook_scraper import get_posts
import pandas as pd
```

```python
#Run this only once waiting time(15Mins)

post_id = []   #creating empty post_id list
post_text = [] #creating empty post_text list
post_time = [] #creating empty post_time list

#Assigning page size randomly to ge the data
#For Page size 200 it is giving data of all the posts from 2019 to 2021
for post in get_posts('walmart', pages=200): 
    post_id.append(post['post_id'])
    post_text.append(post['text'])
    num = post['time']
    date_time = num.strftime("%m/%d/%Y, %H:%M:%S")
    post_time.append(date_time)
```

```python
#Adding all the lists to a dataframe
d = {'Post ID': post_id, 'Post Content': post_text, 'Post Time': post_time}
walmart_df = pd.DataFrame(d)

#Converting the datatype of column Post Time to dattime
walmart_df['Post Time'] = pd.to_datetime(walmart_df['Post Time'])

#Sorting the values
walmart_df.sort_values(by=['Post Time'], inplace=True)

#Reseting the index after sorting the values
walmart_df.reset_index(drop=True,inplace=True)

#Removing all the line breakers to space in all the columns
walmart_df = walmart_df.replace(r'\n',' ', regex=True) 

#Filtering the dataframe to get only the data of 2020 year
filtered_walmart_df = walmart_df.loc[(walmart_df['Post Time'] >= '2020-01-01')
                    & (walmart_df['Post Time'] <= '2020-12-31')]

#Reseting the index after filtering the values
filtered_walmart_df.reset_index(drop=True,inplace=True)


#Checking is there any null values
check_null = filtered_walmart_df.isnull().sum()
print('checking null values')
print(check_null)
print('\n')


#Handeling null values
filtered_walmart_df['Post Content'] = filtered_walmart_df['Post Content'].fillna('NULL')

#Checking is there any null values
print('checking null values after handeling null')
check_null1 = filtered_walmart_df.isnull().sum()
print(check_null1)
print('\n')

filtered_walmart_df
```
